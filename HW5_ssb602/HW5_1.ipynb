### Binomial distribution with mean 100 and 100 samples of size 10

```python
binom_dist = st.binom.rvs(n=200, p=0.5, size=(100,10))
```

```python
pl.figure = figsize(10,10)
pl.hist(binom_dist);
pl.xlabel('Samples')
pl.ylabel('Frequency')
pl.title('Binomial Distribution')
```

### Plotting one sample

```python
def calculate_z(dist,mu=100,std=10):
    z = (np.sqrt(len(dist))*(mu - dist.mean()))/(std)
    return z
```

```python
z_values = np.empty(100)
for i in range(0,100):
    z_values[i] = calculate_z(st.binom.rvs(n=200, p=0.5, size=(100,10)),std=7.07)
```

```python
pl.hist(z_values);
```

### Goodness of fit using AD test

### Null Hypothesis 
$H_0$ = Binomial distribution z-score are consistent with Normal distribution
### Alternate Hypothesis
$H_1$ = Binomial distribution z-score are NOT consistent with Normal distribution
### alpha = 0.05

```python
st.anderson(z_values,'norm')
```

### AD Test Analysis
### Since alpha = 0.05, lookin up significance level array for value 5 and 
### corresponding value in critical value array which is 0.759
### Since Statistic < critical value
### Null Hypothesis cannot be rejected

### Therefore Binomial Distribution z-scores can be said to be consistent
